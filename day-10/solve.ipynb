```python
#!/usr/bin/env python

import re

class Vector(object):
    x = 0
    y = 0
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __mul__(self, other):
        return Vector(self.x * other, self.y * other)
    def __rmul__(self, other):
        return self * other
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

def get_input(fname):
    input = []
    with open(fname) as f:
        for line in f.readlines():
            parts = re.split(r"[=<>, ]+", line.strip())
            input.append((Vector(int(parts[1]), int(parts[2])), Vector(int(parts[4]), int(parts[5]))))
    return input
```

```python
test_input = get_input("test.txt")
```

```python
import math

def print_matrix(matrix):
    print("\n".join(''.join(line) for line in matrix))

def print_messages(input, start=0, loops=100):
    for i in range(start, start+loops):
        xmin, ymin = math.inf, math.inf
        xmax, ymax = -math.inf, -math.inf
        state = []
        for n in input:
            current = n[0] + i * n[1]
            state.append(current)
            xmin = min(current.x, xmin)
            ymin = min(current.y, ymin)
            xmax = max(current.x, xmax)
            ymax = max(current.y, ymax)
        if ymax - ymin + 1 <= 20:
            matrix = [['.' for _ in range(xmax - xmin + 3)] for _ in range(ymax - ymin + 3)]
            for e in state:
                matrix[e.y - ymin + 1][e.x - xmin + 1] = '#'
            print(i)
            print_matrix(matrix)
            print(flush=True)
```

```python
print_messages(test_input, loops=5)
```
